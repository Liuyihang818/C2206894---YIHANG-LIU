```python
import os
import cv2
import numpy as np

from multiprocessing.pool import ThreadPool
Filelist = []

#  the path is where you saved the train  0/7 jpd
for home, dirs, files in os.walk('./train_images'):
    for filename in files:
        # 文件名列表，包含完整路径
        Filelist.append(os.path.join(home, filename))
        
        
# create bag: Each bag consists of 100 images with a fraction x of digit 0 and 1-x of digit 7.
def create_data(item):
    # count is x(the number of digit 0)
    count=0
    for k in range(100):
        randoms = np.random.randint(low=0, high=len(Filelist),size=100)

        for rand in randoms:

            label = int(Filelist[rand].split('.')[1][-1])
            if label == 0:
                count+=1

        img_all = np.zeros((280, 280, 3))
        for i in range(10):
            for j in range(10):
                img1 = cv2.imread(Filelist[randoms[i*10+j]])
                img_all[i*28:i*28+28,j*28:j*28+28,:] = img1

        data = cv2.resize(img_all,dsize=(512,512))
        save_name='./createdata/bag'+str(item+1)+'/'+'bag'+str(item+1)+'_{}.jpg'.format(k)
        if not os.path.exists('./createdata/bag'+str(item+1)):
            os.makedirs('./createdata/bag'+str(item+1))
        cv2.imwrite(save_name, data)
        print(save_name)

    per_0= count / 100
    with open('./createdata/bag'+str(item+1)+'/'+'label.txt', "w") as f:
        f.write(str(per_0))

patch
pool_size = 20

items = list(range(329))

pool = ThreadPool(pool_size)   
pool.map(create_data, items)   
pool.close()  
pool.join()







```
